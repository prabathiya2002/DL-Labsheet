```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Au_nanoparticle_dataset.csv")
df.head()
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
new_df.head(20)
```

```python
mean_values = new_df.mean()
std_values = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean:\n", mean_values)
print("\nStandard Deviation:\n", std_values)
print("\nQuartiles:\n", quartiles)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, col in enumerate(new_df.columns):
    axes[i].hist(new_df[col], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)


g.map_upper(sns.scatterplot)


g.map_diag(sns.histplot, kde=True)


g.map_lower(sns.kdeplot, fill=True, cmap="Blues")

plt.show()
```
